```python
from matplotlib import pyplot as plt
```

```python
from PIL import Image
from PIL import ImageEnhance

# open the image file
image = '/content/drive/MyDrive/Kuliah S2/Semester 1/Computer Vision/content/habib-royni-mujtahid.jpeg'
image = Image.open(image)

# tampilkan gambar
plt.imshow(image)
```

```python
# Meningkatakan kecerahan gambar
current_brightness = ImageEnhance.Brightness(image)
new_brightness = 0.5

# tambah brightness berdasarkan variable new_brightness
img_brightnessed = current_brightness.enhance(new_brightness)

# tampilkan gambar
plt.imshow(img_brightnessed)
```

```python
# Meningkatkan kotras gambar
current_contrast = ImageEnhance.Contrast(image)
new_contrast = 0.5

# tambah contrast berdasarkan variable new_contrast
img_contrast = current_contrast.enhance(new_contrast)

# tampilkan gambar
plt.imshow(img_contrast)
```

```python
# Meningkatkan ketajaman gambar
current_sharp = ImageEnhance.Sharpness(image)
new_sharp = 0.5

# tambah sharp berdasarkan variable new_sharp
img_sharp = current_sharp.enhance(new_sharp)

# tampilkan gambar
plt.imshow(img_sharp)
```

```python
# Mengubah warna
current_color = ImageEnhance.Color(image)
new_color = 5

# Color level enhanced by variable new_color
img_colored = current_color.enhance(new_color)

# shows updated image in image viewer
plt.imshow(img_colored)
```
